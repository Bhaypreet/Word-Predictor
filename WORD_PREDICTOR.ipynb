```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
import numpy as np
import pandas as pd
import string
```

# Lowercase and remove punctuation

```python
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text
```

```python
faqs = preprocess_text(faqs)
```

# Tokenization

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index  
```

```python
num_words = len(tokenizer.word_index)
```

```python
num_words
```

```python
input = []
for sentence in faqs.split('\n'):  
    tokenized_sequence = tokenizer.texts_to_sequences([sentence])[0]  
    for i in range(1, len(tokenized_sequence)):  
        input.append(tokenized_sequence[:i+1])  
```

```python
input
```

# Sequence_Padding

```python
max_len = max([len(x) for x in input])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences = pad_sequences(input,maxlen = max_len,padding = 'pre')
```

```python
padded_input_sequences
```

# Input_Target_Split

```python
x = padded_input_sequences[:,:-1]   #inputs
```

```python
x
```

```python
y = padded_input_sequences[:,-1:]   #targets
```

```python
x.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical 
y = to_categorical(y,num_classes = num_words+1) 
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

## Model Architecture

```python
model1 = Sequential()
```

```python
model1.add(Embedding(input_dim = num_words+1, output_dim=100, input_shape=(max_len,)))

tf.keras.layers.LSTM(128, return_sequences=True),
tf.keras.layers.Dropout(0.2),

tf.keras.layers.LSTM(128, return_sequences=True),
tf.keras.layers.Dropout(0.2),

model1.add(LSTM(150))

model1.add(Dense(num_words+1, activation='softmax'))
```

```python
model1.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model1.build(input_shape=(None, max_len))
model1.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau

reduce_LR = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)
earlystop = EarlyStopping(monitor='loss', patience=10, restore_best_weights=True , verbose=1)
```

## Training Model

```python
model1.fit(x,y,epochs = 50,validation_split = 0.2,batch_size = 32) #callbacks = [earlystop,reduce_LR] 
```

## Testing the Next 5 Words

```python
text = 'What a nice figure'
for i in range(5):
    text_tok = tokenizer.texts_to_sequences([text])[0]
    padded_text = pad_sequences([text_tok],maxlen=max_len,padding = 'pre')
    num = np.argmax(model1.predict(padded_text))  
    for word,index in tokenizer.word_index.items():
        if index == num:
            text = text+" "+word
            print(text)
```
